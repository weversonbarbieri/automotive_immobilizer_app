```python
# List of strings to be found in the column 'Security'
string_keyless_start_list = []
# List of indexes to be found in the column 'Security'
indexes_keyless_start_list = []

# Loop to find the string 'w/ Keyless Start' in the column 'Security'
for index, row in df_mazda.iterrows():
    if 'w/ Keyless Start' in row['Security']:
        string_keyless_start_list.append(row)
        indexes_keyless_start_list.append(index)
```

```python
# Create a DataFrame for Mazda models with 'Keyless Start'
df_mazda_keyless_start = pd.DataFrame(string_keyless_start_list)
```

```python
# Split in 2 lines using the delimeter '\n' of the column 'Security'
df_mazda_keyless_start_line_split = df_mazda_keyless_start.assign(Security=df_mazda_keyless_start['Security'].str.split('\n')).explode('Security')
```

```python
# Create a df containing the string 'Advanced Keyless Entry w/ Keyless Start'
df_mazda_advanced_keyless = df_mazda_keyless_start_line_split[df_mazda_keyless_start_line_split['Security'] == 'Advanced Keyless Entry w/ Keyless Start): M-D']
```

```python
# Replace the string 'MX-5 (Miata)' by 'MX-5 (Advanced Keyless Entry w/ Keyless Start)' in the column 'Model'
df_mazda_advanced_keyless['Model'] = df_mazda_advanced_keyless['Model'].str.replace('MX-5 (Miata)', 'MX-5 (Advanced Keyless Entry w/ Keyless Start)', regex=False)
```

```python
# Replace the string 'Advanced Keyless Entry w/ Keyless Start): M-D' by 'M-D' in the column 'Security'
df_mazda_advanced_keyless['Security'] = df_mazda_advanced_keyless['Security'].str.replace('Advanced Keyless Entry w/ Keyless Start): M-D', 'M-D', regex=False)
```

```python
# Create a df containing the string 'Advanced Keyless Entry w/ Keyless Start'
df_mazda_keyed_ignition = df_mazda_keyless_start_line_split[df_mazda_keyless_start_line_split['Security'] == 'Keyless Entry w/ Keyed Ignition: M-C']
```

```python
# Replace the string 'MX-5 (Miata)' by 'MX-5 (Keyless Entry w/ Keyed Ignition)' in the column 'Model'
df_mazda_keyed_ignition['Model'] = df_mazda_keyed_ignition['Model'].str.replace('MX-5 (Miata)', 'MX-5 (Keyless Entry w/ Keyed Ignition)', regex=False)
```

```python
# Replace the string 'Keyless Entry w/ Keyed Ignition: M-C' by 'M-C' in the column 'Security'
df_mazda_keyed_ignition['Security'] = df_mazda_keyed_ignition['Security'].str.replace('Keyless Entry w/ Keyed Ignition: M-C', 'M-C', regex=False)
```

```python
# Function to replace the strings in the column 'Model' by 'Model (Keyed Ignition)' and 'Model (Advanced Keyless)' based on the string in the column 'Security'
def replace_model_string(row):
    if 'Keyed Ignition' in row['Security']:
        return row['Model'] + ' (Keyed Ignition)'
    elif 'Advanced Keyless' in row['Security']:
        return row['Model'] + ' (Advanced Keyless)'
    return row['Model']
```

```python
# Apply the function 'replace_model_string' to the DataFrame
df_mazda_keyed_ignition_advanced_keyless_line_split['Model'] = df_mazda_keyed_ignition_advanced_keyless_line_split.apply(replace_model_string, axis=1)
```

```python
# Replace the string 'Advanced Keyless: MD' by 'M-D' in the column 'Security'
df_mazda_keyed_ignition_advanced_keyless_line_split['Security'] = df_mazda_keyed_ignition_advanced_keyless_line_split['Security'].str.replace('Advanced Keyless: MD', 'M-D', regex=False)
```

```python
# Create a list of strings for the column 'Security'
current_string_list_security = strings_from_column(df_mazda_keyed_ignition_advanced_keyless_line_split, 'Security')
```

```python
# Function to replace strings based on the declared variables 'current_string' and 'new_string' to be found in the column 'Security'
def replace_security_string(row, current_string, new_string):
    if current_string in row['Security']:
        return row['Security'].replace(current_string, new_string)
    else:
        return row['Security']
```

```python
# First string to be replaced based on the current_string_list_security
current_string = 'Advanced Keyless: M-D'
# New string to replace the first string
new_string = 'M-D'
```

```python
# Apply the function 'replace_security_string' to the DataFrame 'df_mazda_keyed_ignition_advanced_keyless_line_split' 
df_mazda_keyed_ignition_advanced_keyless_line_split['Security'] = df_mazda_keyed_ignition_advanced_keyless_line_split.apply(replace_security_string, axis=1, args=(current_string, new_string))
```
