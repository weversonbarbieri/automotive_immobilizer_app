As the model B9 Tribeca needs to be only in 2 rows (2006 and 2007), the following will be done: 
- Create a dataframe w/ 2 rows for model 2006 and 2007 B9 Tribeca.
- Concatenate this dataframe to the df_subaru.
- Replace the string 'B9 Tribeca (2006 – 2007)\nTribeca (2008 – onward)' for the string 'Tribeca' on the df_subaru
- Drop column 'ParameterReset' as it has no values.

```python
# Create a df for the model B9 Tribeca
df_b9_tribeca = pd.DataFrame({'Year': [2006, 2007], 'Make': ['Subaru', 'Subaru'], 'Model': ['B9 Tribeca', 'B9 Tribeca'], 'Security': ['Std all variants', 'Std all variants']})
```

```python
# Drop column ParameterReset
df_subaru_drop_column = df_subaru.drop(columns=['ParameterReset'])
```

```python
# Replace the string 'B9 Tribeca (2006 – 2007)\nTribeca (2008 – onward)' for the string 'Tribeca'
df_subaru_drop_column.replace(to_replace='B9 Tribeca (2006 – 2007)\nTribeca (2008 – onward)', value='Tribeca', regex=False)
```

```python
# Concatenate df_tribeca to the df_subaru
df_subaru_concatenated = pd.concat([df_b9_tribeca, df_subaru_drop_column])
```

```python
# Create a list w/ the models after updating
subaru_model_list_updated = strings_from_column(df_subaru_concatenated, 'Model')
```
