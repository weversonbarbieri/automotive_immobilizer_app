1. Checking the list w/ all BMW models, these are the ones which need to be changed. As there are 2 models in the same row, they will be separated: 
- 740/745*, 
- 750/760*, 
- 1 Series (128i/135i)

2. The above models will be separated and they should appear as below:
- 740
- 745
- 750
- 760
- 128i
- 135i

The models below do not need to be separated as they are only the type of body:
- M Coupe/Roadster

```python
# Create a list w/ the BMW models containing the caractere '/' 
bmw_model_change_list = []
# Create a list w/ the indexes 
bmw_model_change_index_list = []

# Loop to get all models containing the caractere '/', except model 'M Coupe/Roadster'
for index, row in df_bmw.iterrows():
    if '/' in row['Model'] and row['Model'] != 'M Coupe/Roadster':
        bmw_model_change_list.append(row)
        bmw_model_change_index_list.append(index)
```

```python
# Convert the list of models w/ the caractere '/' to dataframe 
df_bmw_to_change = pd.DataFrame(bmw_model_change_list)
```

```python
df_bmw_to_change
```

```python
# Separate/split the models, using the caractere '/' as delimiter
df_bmw_to_change = df_bmw_to_change.assign(Model=df_bmw_to_change['Model'].str.split('/')).explode('Model')
```

```python
# Create a list to store all rows containing the string and caracteres '1 Series (128i', '135i)', '*'
bmw_remove_char_list = []
# Create a list with all indexes
bmw_remove_char_index_list = []

# Loop into column 'Model', get those rows with the string '1 Series (128i'
for index, row in df_bmw_to_change.iterrows():
    if '1 Series (128i' in row['Model']:
        # Save in the above list
        bmw_remove_char_list.append(row)
        bmw_remove_char_index_list.append(index)

# Change the list to dataframe
df_bmw_remove_char = pd.DataFrame(bmw_remove_char_list)
```

```python
# Remove the string '1 Series (' and keep only the string '128i'.
# The code '...['Model'].replace({"1 Series (": ""}...' generated a error. It was needed to add \\ (... 1 Series \\("...) to fix the error
df_bmw_remove_char['Model'] = df_bmw_remove_char['Model'].replace({"1 Series \\(": ""}, regex=True)
```

```python
# Confirm if the changes were properly done
df_bmw_remove_char.head(2)
```

```python
# Concatenate the main df_bmw to the df_bmw_remove_char
df_bmw_changes_done = pd.concat([df_bmw_to_change, df_bmw_remove_char])
```

```python
# Concatenate the main df_bmw with the df after changes done
df_bmw_concatenated = pd.concat([df_bmw_changes_done, df_bmw])
```

```python
# Reset index
df_bmw_reset_index = df_bmw_concatenated.reset_index(drop=True)
```

```python
# Function to create a list w/ unique values from the columns 
def strings_from_column(df_make, column):
    # List to store the strings from the columns
    strings_from_column_list = []
    # Loop to get the strings from each row based on the column given
    for string in df_make[column]:
        # Store the strings in the list
        strings_from_column_list.append(string)
    # Return a list w/ unique values/strings from the column given
    return set(strings_from_column_list)
```

```python
# Call the function to show a list with unique values containing all strings from column model
bmw_model_unique_string_list_updated = strings_from_column(df_bmw_column_model_first_change, 'Model')
```
